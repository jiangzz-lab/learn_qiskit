```python
from qiskit import *
```

```python
circuit = QuantumCircuit(3, 3)
```

```python
%matplotlib inline
circuit.draw(output='mpl')
```

```python
circuit.x(0)
circuit.barrier()
circuit.draw(output='mpl')
```

```python
circuit.h(1)
circuit.cx(1, 2)
```

```python
circuit.draw(output='mpl')
```

```python
circuit.cx(0, 1)
circuit.h(0)
circuit.draw(output='mpl')
```

```python
circuit.barrier()
circuit.measure([0, 1], [0, 1])
circuit.draw(output='mpl')
```

```python
circuit.barrier()
circuit.cx(1, 2)
circuit.cz(0, 2)
circuit.draw(output='mpl')
```

```python
circuit.measure(2, 2)
simulator = Aer.get_backend('qasm_simulator')
result = execute(circuit, backend=simulator, shots=1024).result()
counts = result.get_counts()
from qiskit.tools.visualization import plot_histogram
plot_histogram(counts)
```

```python
print(counts)
```

```python
# challenge - teleport a superposiotion state on q0 to 02
circuit = QuantumCircuit(3, 3)
circuit.h(0)
circuit.barrier()
circuit.h(1)
circuit.cx(1, 2)
circuit.cx(0, 1)
circuit.h(0)
circuit.barrier()
circuit.measure([0, 1], [0, 1])
circuit.barrier()
circuit.cx(1, 2)
circuit.cz(0, 2)
```

```python
circuit.draw(output='mpl')
```

```python
circuit.measure(2, 2)
```

```python
simulator = Aer.get_backend('qasm_simulator')
result = execute(circuit, backend=simulator, shots=1000).result()
counts = result.get_counts()
%matplotlib inline
plot_histogram(counts)
```
